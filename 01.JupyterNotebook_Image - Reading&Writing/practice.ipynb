## 01. 이미지 읽기

```python
import numpy as np
import cv2
```

### cv2.imread 종류

- cv2.IMREAD_COLOR(): 이미지를 컬러로 로드함, 이미지의 투명한 부분은 모두 무시됨, 디폴트 플래그, 정수값은 1

- cv2.IMREAD_GRAYSCALE(): 흑백 이미지로 로드함. 정수값은 0

- cv2.IMREAD_UNCHANGED: 알파채널을 포함하여 이미지 그대로 로드함. 정수값은 -1

### cv2.waitKey(0)

- 기다리는 시간으로 1/1000초 (ms)

- 0은 키보드 입력값이 있을때까지 대기하라는 말

- 리턴값으로 키보드를 입력한 시간의 초를 반환함

```python
def showImage():
    # 이미지의 경로
    image_path = './me.jpg'
    # imread(보여줄 사진, 읽는 방식(컬러or흑백))
    img = cv2.imread(image_path, cv2.IMREAD_COLOR)
    
    # 객체 표시
    cv2.imshow('myself', img)
    
    # 이미지를 표시한 후 사용자가 키보드를 누를때까지 대기
    cv2.waitKey(0) 
    cv2.destroyAllWindows() # 창 닫음
    
showImage()
    
```

### cv2.namedWindow() & cv2.imshow()

- cv2.WINDOW_AUTOSIZE: 원본 이미지 크기로 고정하여 윈도우를 생성함

- cv2.WINDOW_NORMAL: 원본 이미지 크기로 윈도우를 생성하여 이미지를 나타내지만 사용자가 크기조절 가능!

- *주의* cv2.namedWindow()와 cv2.imshow()의 윈도우 타이틀이 같아야함!! (아래 코드의 경우에는 'myself'로 같은 것을 볼 수 있음)

```python
def showImage():
    image_path = './me.jpg'
    img = cv2.imread(image_path, cv2.IMREAD_COLOR)
    
    cv2.namedWindow('myself',cv2.WINDOW_NORMAL)
    cv2.imshow('myself', img)
    cv2.waitKey(0) 
    cv2.destroyAllWindows() 
    
showImage()
    
```

### imwrite을 통한 이미지 복사!

- waitKey(0)는 사용자가 키보드값을 입력하면, 누른 시간을 반환

- 그러나 반환 값을 wait에 담아서 만약 그 키보드 값이 'c'일 경우 - 사진 복사!

- *중요* 여기서 ord('c')는 문자 c의 ASKII CODE 값 99를 리턴함

- ord() <-> chr(), #chr(99) == 'c'

```python
def showImage():
    image_path = './me.jpg'
    img = cv2.imread(image_path, cv2.IMREAD_COLOR)
    cv2.imshow('myself', img)
    
    wait = cv2.waitKey(0) & 0xFF # 0xFF는 64->32bit 운영 체제 때문에! 
    
    if wait == 27:
        cv2.destroyAllWindows() 
    elif wait == ord('c'):
        cv2.imwrite('./myself_copy.jpg', img)
        cv2.destroyAllWindows()    
    
    
    
showImage()
    
```

### Matplotlib.pyplot을 통한 이미지 표시

```python
%matplotlib inline
import matplotlib.pyplot as plt 

def showImage():
    image_path = './me.jpg'
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE) # 이미지를 흑백으로 읽음 - IMREAD_GRAYSCALE
    plt.imshow(img, cmap='gray', interpolation='bicubic') # imshow함수를 이용해 이미지를 디스플레이함
    
    # 이미지를 눈금 없이 표
    plt.xticks([]) 
    plt.yticks([])
    plt.title('Myself')
    plt.show()
    
showImage()
```
